### Analysis

25 Mice were the subjest of a study of the effects of Capomulin on squamous cell carcinoma (SCC) tumor sizes. Capomulin was added as a treatment regime on top of 9 existing treatments. The results of the Capomulin trials have been thoroughly analysed to provide some insight into the results and to assist making a decision on which drugs to continue into the next phase of testing.

Mice treated with Capomulin ranked first in smallest median tumor volume across the entire study, with the lowest variance in results. 92% of Mice treated with Capomulin saw reductions in tumor sizes in their final measurement. This indicates Campomulin is effective at treating SCC in mice more consistently when compared to the other drugs in the trial. Analysing the correlation between mouse weight (g) and average tumor size also indicated that Capomulin is slightly more effective at treating heaver mice. 



An existing drug in the trial, Ramicane, also demonstrated results incredibly similar to Campomulin. It ranked second in median tumor volume, and had an almost identical variance to Campomulin. Ramicane was the second most consistently monitored drug, with more mice being monitored for the entire 45 day period. Pymaceuticals should consider the cost, availability and treatment medium when deciding which drugs move onto the next phase of trial. Ramicane may well prove to be a better alternative when factoring in all these variables.

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
from sklearn import datasets
from scipy.stats import linregress

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single DataFrame
df = pd.merge(study_results, mouse_metadata, how = 'inner')
# Display the data table for preview
df.head()
```

```python
# filter by the one time point to get a df of all mouse id's 
df_sliced = df.loc[df['Timepoint'] == 0]
df_sliced
```

```python
df_sliced['Drug Regimen'].value_counts()
```

```python
# Checking the number of mice.
mice_count = len(df['Mouse ID'].unique())
mice_count
```

```python
# the lambda function is used to loop through each unique id in the dataframe and return only those whose counts are > 1, i.e the duplicate mice.
duplicate_id = df_sliced.groupby('Mouse ID').filter(lambda x: len(x) > 1)['Mouse ID'].unique()
duplicate_id
```

```python
# Our data should be uniquely identified by Mouse ID and Timepoint
# Using the isin function and the to ask for all information that is in the duplicate_id variable and only showing the id and timepoint variable.
df_sliced.loc[df_sliced['Mouse ID'].isin(duplicate_id), ['Mouse ID', 'Timepoint']]
```

```python
# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
# Optional: Get all the data for the duplicate mouse ID. 
df.loc[df['Mouse ID'].isin(duplicate_id)]
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID. Using the isin function and the ~ at the start to ask for all information that is not in the 
# duplicate_id variable
df_cleaned = df.loc[~df['Mouse ID'].isin(duplicate_id)]

df_cleaned
```

```python
# Checking the number of mice in the clean DataFrame.
mouse_count_cleaned = len(df_cleaned['Mouse ID'].unique())
mouse_count_cleaned
```

## Summary Statistics

```python
tumor_mean = df_cleaned.groupby('Drug Regimen')['Tumor Volume (mm3)'].mean()
tumor_mean

tumor_median = df_cleaned.groupby('Drug Regimen')['Tumor Volume (mm3)'].median()
tumor_median

tumor_var = df_cleaned.groupby('Drug Regimen')['Tumor Volume (mm3)'].var()
tumor_var

tumor_std = df_cleaned.groupby('Drug Regimen')['Tumor Volume (mm3)'].std()
tumor_std

tumor_sem = df_cleaned.groupby('Drug Regimen')['Tumor Volume (mm3)'].sem()
tumor_sem

# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary DataFrame.
drug_summary_df = pd.DataFrame({'mean Tumor Volume' : tumor_mean,
                                'median Tumor Volume (mm3)' : tumor_median,
                                'Tumor Volume variance' : tumor_var,
                                'Tumor Volume standard deviation' : tumor_std,
                                'Tumor Volume standard error mean' : tumor_sem}
                              )
drug_summary_df
```

```python
# A more advanced method to generate a summary statistics table of mean, median, variance, standard deviation,
# and SEM of the tumor volume for each regimen (only one method is required in the solution)

# Using the aggregation method, produce the same summary statistics in a single line
agg_df = df_cleaned.groupby('Drug Regimen')['Tumor Volume (mm3)'].agg(['mean', 
                                                                       'median', 
                                                                       'var', 
                                                                       'std', 
                                                                       'sem']
                                                                     )
agg_df
```

## Quartiles, Outliers and Boxplots

```python
# Capomulin, Ramicane, Infubinol, and Ceftamin
sample_Drugs = ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']

# Slicing the data to include results that only contain the drugs in the sample_drugs variable
time_point_df = df_cleaned.loc[(df_cleaned['Drug Regimen'].isin(sample_Drugs))]
time_point_df.head()
```

```python
# Get the mouse ids at the final timepoint. Calling the index because GroupBy automatically assigns the called group as the index.
final_vol_ids = time_point_df.groupby('Mouse ID')['Timepoint'].max().index

# Start by getting the last (greatest) timepoint for each mouse
# Get the final time point for each mouse using the .max() command. Call the values because we only want to save them as the final_timepoints variable for use in the dataframe.
final_timepoints = time_point_df.groupby('Mouse ID')['Timepoint'].max().values

# Calculate the final tumor volume of each mouse across four of the treatment regimens:
# Getting the final tumor volume for each mouse & timepoint in the above two variables and saving that in a list.
final_tumor_vol = []
index = np.arange(0, len(final_vol_ids))

# Loop through each slice matching the mouse ID & final timepoint for said ID to find the final volume. Store the value as a float in the variable so it integrates into the final DF properly.
for i in index:
    final_vol = df_cleaned.loc[(df_cleaned['Mouse ID'] == final_vol_ids[i]) & (df_cleaned['Timepoint'] == final_timepoints[i]), ['Tumor Volume (mm3)']]
    final_vol_as_float = float(final_vol.values)
    
    final_tumor_vol.append(final_vol_as_float)

final_id_timepoints = pd.DataFrame({ 'Mouse ID' : final_vol_ids,
                                   'Timepoint' : final_timepoints, 
                                    'Final Tumor Volume (mm3)' : final_tumor_vol
                                   })
final_id_timepoints
```

```python
# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
# Merge on timepoints and mouse id to only include the mice and their respective final measured timepoints.
combined_df = pd.merge(df_cleaned, final_id_timepoints, on = ['Mouse ID', 'Timepoint'])

combined_df
```

```python
# Put treatments into a list for for loop (and later for plot labels)
drug_list = ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']
# Because the groupby functions generally all act on sorted data, sorting this list now ensures output data is consistent later.
drug_list.sort()
drug_list
```

```python
# Create empty list to fill with tumor vol data (for plotting)
tumor_data = []

# Used the .apply function to run the function group_tumor_data
# The group_tumor_data function takes the information for each group of Drug Regimen & Final Tumor Volume Data in the groupby and appends it to the tumor_data list as a complete set of data
def group_tumor_data(drug):
    tumor_data.append(drug.values)
    

combined_df.groupby('Drug Regimen')['Final Tumor Volume (mm3)'].apply(group_tumor_data)

tumor_data
```

```python
# Generate a box plot that shows the distrubution of the tumor volume for each treatment group.
fig1, ax1 = plt.subplots()
ax1.set_title('Analysis of Final Tumor Size for Each Drug Regime')
ax1.set_ylabel('Tumor size (mm3)')
ax1.boxplot(tumor_data, labels = drug_list, sym = 'rD')

plt.show()
```

```python
# Created a list within lists to store each of the drugs information in each sub-list to call on later for calculations
Drug_Summary = [ [] for i in range(len(drug_list))]
Drug_Summary

drug_counter = range(len(drug_list))

# Created a counter to increase by 1 each iteration to call on to append data to each list within list in the Drug_Summary bucket
count = 0

drug_list

# Calculate the IQR and quantitatively determine if there are any potential outliers.     
# Locate the rows which contain mice on each drug and get the tumor volumes    
# add subset     
# Determine outliers using upper and lower bounds

for i in drug_counter:
    
# Loop goes through the combined dataframe (with the final tomor volumes of each mouse) and slices the data to mach with each drug in drug list and determines the quartiles [0 to 2] and stores the values
    quartiles = combined_df.loc[combined_df['Drug Regimen'] == drug_list[i],['Final Tumor Volume (mm3)']].quantile([.25,.5,.75]).values
    
# values are stored in variables as floats to calculate with later
    lowerq = float(quartiles[0])
    upperq = float(quartiles[2])
    iqr = float(quartiles[2] - quartiles[0])
    
# Slice the data further to search for outliers in the table +- 1.5 * iqr for the current drug it's looping through.

    upper_outliers = combined_df.loc[(combined_df['Final Tumor Volume (mm3)'].values > (upperq + (1.5 * iqr))) & (combined_df['Drug Regimen'] == drug_list[i])]
    lower_outliers = combined_df.loc[(combined_df['Final Tumor Volume (mm3)'].values < (lowerq - (1.5 * iqr))) & (combined_df['Drug Regimen'] == drug_list[i])]

# append the information to the Drug_Summary[count] bucket. i.e Drug_Summary[0] on the first iteration, Drug_Summary[1] on the next.
    Drug_Summary[count].append(lowerq)
    Drug_Summary[count].append(upperq)
    Drug_Summary[count].append(iqr)

# Appends outliers both upper and lower to the buckets if there are outliers found in the previous steps.
    if len(upper_outliers) > 0:
        Drug_Summary[count].append(upper_outliers[['Mouse ID', 'Final Tumor Volume (mm3)']])

    if len(lower_outliers) > 0:
        Drug_Summary[count].append(lower_outliers[['Mouse ID', 'Final Tumor Volume (mm3)']])

# Increase the counter by 1 to loop to the next Drug_Summary bucket       
    count =  count + 1

# Store a variable which is the length of the bucket we stored information in. If it's length 3, there are no outliers, If 4 or 5 there are upper & lower outliers respectively.
    drug_len = len(Drug_Summary[i])
    
    print(f'The IQR of {drug_list[i]} is {Drug_Summary[i][2]}')
    
    if (drug_len == 4):
        print(f'The outliers of {drug_list[i]} are {Drug_Summary[i][3]}')
        
    if (drug_len == 5):
         print(f'The outliers of {drug_list[i]} are {Drug_Summary[i][3]} {Drug_Summary[i][4]}')
```

```python
# Generate a scatter plot of mouse weight vs. the average observed tumor volume for the entire Capomulin regimen
capomulin_df = df_cleaned.loc[df_cleaned['Drug Regimen'] == 'Capomulin']
capomulin_df

# Groupby  the mouse ID & weights, and drop the weights to a column using reset_index
# Get the values of each weight to use as the x axis values
weight_each = (capomulin_df.groupby(['Mouse ID', 'Weight (g)'])['Tumor Volume (mm3)'].mean()).reset_index('Weight (g)')
weight_each = weight_each['Weight (g)'].values
weight_each

# Get the values of tunor size (mean) to use as the y axis values
vol_mm3 = (capomulin_df.groupby(['Mouse ID', 'Weight (g)'])['Tumor Volume (mm3)'].mean()).reset_index('Weight (g)')
vol_mm3 = vol_mm3['Tumor Volume (mm3)'].values
vol_mm3

plt.scatter(weight_each, vol_mm3, marker = 'x', color = 'purple')
plt.ylabel('Average Tumor Size (mm3)')
plt.xlabel('Mouse Weight (g)')
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and a linear regression model 
# for mouse weight and average observed tumor volume for the entire Capomulin regimen

pearson_coefficient = st.pearsonr(weight_each,vol_mm3)
print(f"The correlation between both factors is {round(pearson_coefficient[0],2)}")
```

```python
x_values = weight_each
y_values = vol_mm3
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(0,50),fontsize=15,color="red")
plt.xlabel('Mouse Weight (g)')
plt.ylabel('Average Tumor Size (mm3)')
print(f"The r-squared is: {rvalue**2}")
plt.show()
```
